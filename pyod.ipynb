```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

def data_generation(_type):
    n_samples = 10081
    anomalies = np.random.choice(np.arange(n_samples), 5)
    if _type == 1:
        arr = np.random.normal(0, 1, size = (n_samples, ))
        for idx in anomalies:
            arr[idx] = 5
    if _type == 2:
        arr = np.random.normal(0, 1, size = (n_samples, ))
        mask = (np.random.normal(size = (10081, )) > 2).astype(int)
        arr *= mask
        arr += np.random.normal(0, 0.01, size = arr.shape)
        for idx in anomalies:
            arr[idx] = 5
    return arr, anomalies
```

# pyod 使用

```python
from pyod.models.ecod import ECOD
data, label = data_generation(2)
clf = ECOD()
x = data.reshape(-1, 1)
clf.fit(x)
y_pred = clf.predict(x)
y_scores = clf.decision_function(x)

plt.figure(figsize = (24, 3))
plt.plot(x, color = "blue")
plt.plot(y_scores, color = "red", alpha = .4)
plt.show()
```
